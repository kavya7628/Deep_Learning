```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/Users/kkutumba/Downloads/restaurant_data.csv")
df.head()
```

```python
df.dtypes
```

```python
df.drop('Name',axis='columns',inplace=True)
```

```python
df.drop('Cuisine',axis='columns',inplace=True)
df.head()
```

```python
df.shape
```

```python
for Location in df:
    print(f'{Location} : {df[Location].unique()}')
```

```python
df['Parking Availability'].replace({'Yes':'1','No':'0'}, inplace=True)
```

```python
df1 = pd.get_dummies(data=df, columns=['Location'])
```

```python
df1.head()
```

```python
cols_to_change = ['Location_Downtown','Location_Rural','Location_Suburban','Parking Availability']
df1[cols_to_change] = df1[cols_to_change].astype(bool).astype(int)
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
column_to_scale = ['Rating','Seating Capacity','Average Meal Price','Marketing Budget','Social Media Followers','Chef Experience Years','Number of Reviews','Avg Review Length','Ambience Score','Service Quality Score','Weekend Reservations','Weekday Reservations','Revenue']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[column_to_scale] = scaler.fit_transform(df1[column_to_scale])
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
X = df1.drop('Revenue', axis='columns')
y = df1['Revenue']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([
    keras.layers.Dense(16, input_shape=(16,), activation='relu'),
    keras.layers.Dense(1),
])
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mean_squared_error'])
model.fit(X_train, y_train, epochs=49)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:4]
```

```python
y_test[:4]
```

```python
ypred = []
for i in yp:
    if i<0.5:
        ypred.append(0)
    else:
        ypred.append(1)
        
```

```python
ypred[:4]
```

```python
ytest = []
for j in y_test:
    if j<0.5:
        ytest.append(0)
    else:
        ytest.append(1)
        
```

```python
ytest[:4]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(ytest, ypred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=ytest, predictions=ypred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
